```python
#Data Cleaning and Preparation
#handles missing values by replacing NaNs with the mean value of the respective columns and then removing any remaining rows with missing values.
```

```python
data1=df.replace({'Status':{'Developing':0,'Developed':1}})
data1
```

```python
data1.isnull().sum(axis = 0)
```

```python
data1['Population'].mean()
```

```python
data1['Population'] = data1['Population'].replace(np.NaN, data1['Population'].mean())
data1['Population']
```

```python
data1.isnull().sum(axis = 0)
```

```python
data1['GDP'].mean()
data1['GDP'] = data1['GDP'].replace(np.NaN, data1['GDP'].mean())
data1['GDP']
data1.isnull().sum(axis = 0)
```

```python
data1['Total expenditure'].mean()
data1['Total expenditure'] = data1['Total expenditure'].replace(np.NaN, data1['Total expenditure'].mean())
data1['Total expenditure']
data1.isnull().sum(axis = 0)
```

```python
data1['Hepatitis B'].mean()
data1['Hepatitis B'] = data1['Hepatitis B'].replace(np.NaN, data1['Hepatitis B'].mean())
data1['Hepatitis B']
data1.isnull().sum(axis = 0)
```

```python
data1.dropna(inplace=True)
data1
```

```python
data1.isnull().sum()
```

```python
x=data1.drop(['Life expectancy ','Country'],axis=1).values
```

```python
y=data1['Life expectancy '].values.reshape(-1,1)
```

```python
#Splitting the Data, splits the dataset into features (x) and target variable (y), and then further splits the data into training and testing sets.
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
#initializes a linear regression model and trains it using the training data.
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
#Model Evaluation, evaluates the model's performance by printing the shapes of the training and testing data and 
#calculating the model's accuracy scores on both.
```

```python
model.score(x_test,y_test)
```

```python
#Making Predictions
#makes predictions on the training and testing data, then calculates and prints the mean absolute error and mean squared error for both.
```

```python
y_p_train=model.predict(x_train)
```

```python
y_p_test=model.predict(x_test)
```

```python
print("Mean Absolute Error: " , mean_absolute_error(y_train,y_p_train))
print("Mean Absolute Error: " , mean_absolute_error(y_test,y_p_test))   
print("Mean Squared Error: " , mean_squared_error(y_train,y_p_train)) 
print("Mean Squared Error: " , mean_squared_error(y_test,y_p_test))   
```

```python
#Cross-Validation
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
kf = KFold(n_splits=8, shuffle=True, random_state = 1)

cvs_model = cross_val_score(model, x, y, cv=kf, scoring='r2')
print(cvs_model)
```
